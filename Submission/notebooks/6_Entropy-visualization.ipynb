```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from math import log as logarithm

def binary_entropy(x):
    if x in {0, 1}:
        return 0
    else:
        return x * logarithm(1. / x, 2) + (1 - x) * logarithm(1. / (1 - x), 2)
```

```python
def binary_misclassification(x):
    return min(x, 1 - x)
```

```python
x = np.linspace(0, 1, num=100)
```

```python
y_entropy = [binary_entropy(xi) for xi in x]
y_misclassification_scaled = [2 * binary_misclassification(xi) for xi in x]
```

```python
label_a, = plt.plot(x, y_entropy, label="entropy")
label_b, = plt.plot(x, y_misclassification_scaled, label="misclassification (scaled)")
plt.legend(handles=[label_a, label_b], bbox_to_anchor=(1.65, 1))
plt.xlabel("probability of first class")
plt.ylabel("uncertainty")
plt.show()
```
